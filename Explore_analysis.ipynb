```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('usa_mercedes_benz_prices.csv')
df.head(5)
```

```python
df.info()
```

```python
df = df[df['Price'] != 'Not Priced']
```

```python
df['Price'] = df['Price'].astype(float)
```

```python
sns.pairplot(df)
plt.show()
```

```python
df['Rating'] = df['Rating'].astype(object)
```

```python
df['Mileage'] = df['Mileage'].str.replace('mi','').str.replace(',','.')
```

```python
df['Mileage'] = df['Mileage'].str.replace(' ','')
```

```python
df['Mileage'] = df['Mileage'].astype(str).str[:-1].astype(float)
```

```python
df.Rating = df.Rating.fillna(df.Rating.mode()[0])
```

```python
df.info()
```

```python
df.info()
```

```python
df['Rating'] = df['Rating'].astype(object)
```

```python
df['Year'] = df['Name'].apply(lambda x: x.split()[0])
```

```python
df.info()
```

```python
df['Review Count'] = df['Review Count'].fillna(df.Rating.mode()[0])
```

```python
df['Review Count'] = df['Review Count'].astype(int)
```

```python
sns.scatterplot(df,palette='viridis',x='Rating',y='Price')
plt.xlabel('Mileage',fontsize=12)
plt.ylabel('Price.10³',fontsize=12)
plt.show()
```
